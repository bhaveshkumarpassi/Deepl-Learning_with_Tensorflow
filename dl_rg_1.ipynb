```python
X = df.drop('price',axis=1).values
y = df['price'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
 X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X_train.shape
```

```python
model = Sequential()

# roughly we can get an idea of neurons in a layer and no. of layers from the shape of train set another way is early stopping
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))

model.add(Dense(1))

model.compile(optimizer='adam',loss='mse')
```

```python
model.fit(X_train,y_train,validation_data=(X_test,y_test),
         batch_size=128,epochs=600) 
# validation data added is nothing like leaking data it will not contibute towards parameter updates by an optimizer but
# it's just added so we can keep a check of how our model is performing on train set and test set as well
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

```python
from sklearn import metrics
```

```python
predictions = model.predict(X_test)
```

```python
np.sqrt(metrics.mean_squared_error(y_test,predictions))
```

```python
metrics.mean_absolute_error(y_test,predictions)
```

```python
metrics.explained_variance_score(y_test,predictions)
```

```python
metrics.explained_variance_score(y_train,model.predict(X_train))
```

```python
plt.figure(figsize=(10,6))
plt.scatter(y_test,predictions)
plt.plot(y_test,y_test,'r--')
```
